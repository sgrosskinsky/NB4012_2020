```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import sdeint
```

## Wright-Fisher diffusion

Simulate $dX_t =\sqrt{2X_t (1-X_t)}dB_t$

```python
tspan = np.linspace(0.0, 5.0, 5001)

def a(x, t):
    return 0.

def sig(x, t):
#    return np.sqrt(2*x*(1-x))
    return 2*x*(1-x)

for n in range(10):
    result = sdeint.itoint(a, sig, 0.5, tspan)
    plt.plot(tspan,result)

plt.title('Wright-Fisher diffusion')
plt.xlabel('$time\ t$')
plt.ylabel('$X_t$')
```

## Geometric Brownian motion

Simulate $dY_t =(\mu +\sigma^2 /2) Y_t dt+\sigma Y_t dB_t$

```python
mu=-0.25
ss=1

#tspan = np.arange(0,tmax+deltat,deltat)/L
tspan = np.linspace(0.0, 5.0, 501)

def a(x, t):
    return x*(mu+ss/2)

def sig(x, t):
    return x*np.sqrt(ss)

plt.figure(figsize = (20,8))

for n in range(10):
    result = sdeint.itoint(a, sig, 1., tspan)
    plt.subplot(1,2,1)
    plt.plot(tspan,result)
    plt.yscale('linear')
    
    plt.subplot(1,2,2)
    plt.plot(tspan,result)
    plt.yscale('log')

# Theory

f = np.exp( (mu +ss/2)*tspan )
plt.subplot(1,2,1)
plt.title('Geometric Brownian motion')
plt.plot(tspan,f,'k')
plt.xlabel('$time\ t$')
plt.ylabel('$Y_t$')
plt.yscale('linear')

plt.subplot(1,2,2)
plt.title('logarithmic scale')
plt.plot(tspan,f,'k')
plt.xlabel('$time\ t$')
plt.ylabel('$Y_t$')
plt.yscale('log')
```

Simulate $X_t =\mu t-\sigma B_t$ and then plot $Y_t =e^{X_t}$

```python
mu=-0.25
ss=1

#tspan = np.arange(0,tmax+deltat,deltat)/L
tspan = np.linspace(0.0, 5.0, 501)

def a(x, t):
    return mu

def sig(x, t):
    return np.sqrt(ss)

for n in range(10):
    result = sdeint.itoint(a, sig, 1., tspan)
    plt.plot(tspan,np.exp(result))
    plt.yscale('linear')

# Theory

f = np.exp( (mu +ss/2)*tspan )
plt.title('Geometric Brownian motion')
plt.plot(tspan,f,'k')
plt.xlabel('$time\ t$')
plt.ylabel('$Y_t$')
plt.yscale('linear')
```
